```python
import sqlite3
import pandas as pd
import numpy as np
```

```python
conn = sqlite3.connect('scorecard.db', detect_types=sqlite3.PARSE_COLNAMES)
cur = conn.cursor()
```

```python
cur.execute('''SELECT * FROM twoyear;''')
df2 = pd.DataFrame(cur.fetchall())
df2.columns = [i[0] for i in cur.description]
df2.head()
```

```python
df2['completion_rate__less_than_4yr_150nt'] = df2['completion_rate__less_than_4yr_150nt'].replace(-999, np.nan)
df2 = df2.dropna(axis=0, subset=['completion_rate__less_than_4yr_150nt'])
```

```python
df2.drop(labels=['parents_middleschool','parents_highschool', 'parents_somecollege'], axis=1, inplace=True)
```

```python
df2 = df2.replace(-999, np.nan)
```

```python
df2.drop(labels='retention_rate_part_time', axis=1, inplace=True)
```

```python
df2.isna().sum(axis=1).sort_values(ascending=False)
```

```python
df2.drop(df2[df2.isna().sum(axis=1)>3].index, inplace=True)
```

```python
df2.isna().sum()
```

```python
df2.iloc[:,2:] = df2.iloc[:,2:].apply(lambda x: x.fillna(x.median()), axis=0)
```

```python
df2.isna().sum()
```

```python
df2.describe()
```

```python
df2['size']=df2['size'].apply(lambda x: (x - df2['size'].min())/df2['size'].max())
df2['cost']=df2['cost'].apply(lambda x: (x - df2['cost'].min())/df2['cost'].max())
df2['median_hh_income']=df2['median_hh_income'].apply(lambda x: (x - df2['median_hh_income'].min())/df2['median_hh_income'].max())
```

```python
def remove_outlier(df_in, col_name):
    q1 = df_in[col_name].quantile(0.25)
    q3 = df_in[col_name].quantile(0.75)
    iqr=q3 -q1
    fence_low= q1-1.5*iqr
    fence_high= q3+1.5*iqr
    df_out=df_in.loc[(df_in[col_name]>fence_low) & (df_in[col_name]<fence_high)]
    return df_out 
```

```python
print(len(df2))
for feature in df2.columns[2:]:
    df2=remove_outlier(df2,feature)
print(len(df2))
```

```python
df2['pellsq']=(df2['pell_grant_rate'])**2
df2['incost']=df2['median_hh_income']*df2['cost']
```
